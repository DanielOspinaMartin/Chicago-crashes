```python
import pandas as pd
chicago_data = 'https://raw.githubusercontent.com/terranigmark/curso-analisis-exploratorio-datos-platzi/main/Traffic_Crashes1.csv'
df_chicago = pd.read_csv(chicago_data,error_bad_lines=False)
```

```python
df_chicago.info()
```

**I decided to see which streets had the most crashes with injuries or tow due to crashs and without injuries**

That is why I filtered the data set in one data set with all the crashes with injuries or tow due to crashs and another one with the the data of all the crahes without injuries


```python
df_chicagof1 = df_chicago[df_chicago['CRASH_TYPE'] == 'INJURY AND / OR TOW DUE TO CRASH']
df_chicagof2 = df_chicago[df_chicago['CRASH_TYPE'] == 'NO INJURY / DRIVE AWAY']
```

```python
df_chicagof1.head()
```

```python
df_chicagof2.head()
```

**Which street has the biggest quantity of crashes with injuries or tow due to crash?.**

```python
tab6 = pd.crosstab(index =df_chicagof1['STREET_NAME'],columns='frecuencia INJURY AND / OR TOW DUE TO CRASH' ).sort_values('frecuencia INJURY AND / OR TOW DUE TO CRASH',axis=0)
tab6
```

```python
tab6['frecuencia INJURY AND / OR TOW DUE TO CRASH'].sum()
```

```python
tab6_1 = tab6[tab6['frecuencia INJURY AND / OR TOW DUE TO CRASH'] >= 7]
tab6_1.sum()
```

```python
tab7 = pd.crosstab(index =df_chicagof2['STREET_NAME'],columns='frecuencia NO INJURY / DRIVE AWAY' ).sort_values('frecuencia NO INJURY / DRIVE AWAY',axis=0)
tab7
```

```python
tab7_1 = tab7[tab7['frecuencia NO INJURY / DRIVE AWAY'] >= 16]
tab7_1
```

```python
tab7_1.sum()
```
